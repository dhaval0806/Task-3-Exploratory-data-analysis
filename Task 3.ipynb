# Clean the data

```python
dataset1=dataset.drop(columns=['Postal Code'], axis=1)
print(dataset1)
```

```python
dataset1.corr()
```

# Data visulization

```python
plt.figure(figsize=(25,8))
plt.bar('Sub-Category','Category',data = dataset1)
plt.title("Graph of Sub_category vs Category")
plt.xlabel('Sub-Category')
plt.ylabel('Category')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(25,8))
plt.bar('Category','Sales',data = dataset1)
plt.title("Graph of Sales vs Category")
plt.xlabel('Category')
plt.ylabel('Sales')
plt.show()
```

So, above plotted graph shows that technology category has more sales over any other category. and furniture has very less sale so, there is a need of good marketing stratergy in furniture category.

```python
plt.figure(figsize=(25,8))
plt.bar('Category','Profit',data = dataset1)
plt.title("Graph of Profit vs Category")
plt.xlabel('Category')
plt.ylabel('Profit')
plt.show()
```

```python
plt.figure(figsize=(25,8))
plt.bar('Sub-Category','Sales',data = dataset1)
plt.title("Graph of Sales vs Sub_Category")
plt.xlabel('Sub-Category')
plt.ylabel('Sales')
plt.show()
```

```python
plt.figure(figsize=(25,8))
plt.bar('Region','Sales',data = dataset1)
plt.title("Graph of Sales vs Region")
plt.xlabel('Region')
plt.ylabel('Sales')
plt.show()
```

So, as seen in above graph EAST region has a lesser sale than any other region

```python
plt.figure(figsize=(25,8))
plt.bar('Region','Profit',data = dataset1)
plt.title("Graph of Profit vs Region")
plt.xlabel('Region')
plt.ylabel('Profit')
plt.show()
```

# Statewise analysis

```python
plt.figure(figsize=(15,12))
sns.countplot(x='State', data=dataset, palette='rocket_r',order=dataset['State'].value_counts().index)
plt.xticks(rotation=90)
plt.show()
```

So, here we conclude that california has highest number of order recieved and it is most sctive state for the selling purpose

# Profit and discount relation

```python
sns.lineplot(x='Discount',y='Profit',label='Profit',data=dataset)
plt.legend()
plt.show()
```

So, we can conclude that at point 5 discount profit is minimum

# Profit and quantity relation

```python
sns.lineplot(x='Quantity',y='Profit',label='Profit',data=dataset)
plt.legend()
plt.show()
```

So, if quantity is high then profit will also high

# Profit/loss and sales by states

```python
ds=dataset.groupby('State')[['Sales','Profit']].sum().sort_values(by='Sales', ascending=False)
ds[:].plot.bar(color=['Blue','Red'],figsize=(15,12))
plt.title("Profit/loss and sales across the state")
plt.xlabel('States')
plt.ylabel('Profit/loss and sales')
plt.show()
```

# Profit/loss and sales by Sub-category

```python
ds=dataset.groupby('Sub-Category')[['Sales','Profit']].sum().sort_values(by='Sub-Category', ascending=False)
ds[:].plot.bar(color=['Blue','Red'],figsize=(15,12))
plt.title("Profit/loss and sales across the Sub-category")
plt.xlabel('Sub-Category')
plt.ylabel('Profit/loss and sales')
plt.show()
```
